##Linear Regression

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
```

```python
## Fetching the Life expectancy dataset that is ready for modelling
```

```python
df1=pd.read_csv('/content/drive/MyDrive/ColabNotebooks/finalmerged.csv')
df=df1.iloc[:,1:]
df
```

```python
df.isnull().sum()   
```

```python
df.fillna(0, inplace=True)    # 1 Attribute with Nulls, Vaccination for 8 records can be replaced with 0.
```

```python
x=df.drop(['Lifeexpectancy'], axis=1)
y=df['Lifeexpectancy']
```

```python
# Object type attributes are wrangled for modeling using get_dummies method.
```

```python
x = pd.get_dummies(data=x, drop_first=True)     
x
```

```python
#Train-Test Split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
#Applying
```

```python
from sklearn.linear_model import LinearRegression
reg_model = LinearRegression()
reg_model.fit(x, y)
y_predicted = reg_model.predict(x_test)
y_predicted
```

```python
reg_model.score(x_test,y_test)     #Accuracy score   96.8%
```

```python
viz=df1.iloc[x_test.index]
viz1=viz['Lifeexpectancy']
viz1
```

```python
# Predicted Label values of 20% tested data vs Original Label values.
```

```python
vizdf=pd.DataFrame(data=viz1)
vizdf['y_Predicted']=y_predicted
vizdf
```

```python
print('mean_squared_error : ', mean_squared_error(y_test, y_predicted))
print('mean_absolute_error : ', mean_absolute_error(y_test, y_predicted))
```
